```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies['cast']
```

find only those dictionaries where the value of job is director. searching this is difficult. so create a function for this.

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies['crew']
```

'overview' column is a string, make it a list in order to concatenate it with other lists

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(3)
```

there's also a person called Sam Mendes. so, when tags are created, 'Sam' tag will be created. when you ask for Sam Mendes movies, Sam Worthington's movies may be recommended. Hence, convert 'Sam Worthington' to 'SamWorthington'.

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']
```

```python
def collapse(words):
    L1 = []
    for i in words:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

To calculate similarity between the Movies, you need to calculate a similarity between the above text tags. Use similarity score for this. This is very difficult to do in terms of texts. So what do you need to do is convert this text into vectors, use X and Y coordinates in 2D space.

Each movie is a vector now. so you need to catch the nearest vectors. This process is called text vectorization.

Text vectorization has many techniques such as bag of words, tfidf, word2vec etc. Here, we will use bagofwords.

Concatenate each tag to form a large text. Find 5000 most common words appearing in the text (Calculate the frequency of the words). Form table using this, with 'movies' as rows and words as columns where the words appear in each movie. Enter the frequency in the table.

The increase in the number of words will lead to the increase in the dimensions. So choose as many less number of words as possible. Try with different dimensions.

Remove stopwords before performing vectorization.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
new_df['tags']
```

Stemming is now done, so let's go back and start the vectorization process again.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new_df['tags']).toarray()
vector[0]
```

Now we need to find the distance between the vectors. (For higher dimensions, Euclidean distance is not a good measure. It's not a reliable one.) We shall use cosine function and find the degrees distance between two vectors. More the cosine distance, the less similar are the movies.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

Fetch the index, sort the nearest distances.

Sorting will make you lose index position. So use enumerate function to have tuples with index position and values.

```python
# sort on the basis of 2nd one in the tuple 
sorted(list(enumerate(similarity[0])), key=lambda x:x[1])
```

```python
# sort in reverse order to get the most similar ones in the top
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])
```

```python
# as the first movie is itself, we shall consider the next five movies
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
new_df.iloc[1216].title
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]    
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```

```python
recommend('Batman Begins')
```
